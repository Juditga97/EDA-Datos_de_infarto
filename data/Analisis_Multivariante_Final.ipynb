```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Heart_2022_Cleaned.csv")
df.head(5)
```

```python
cols = [
    "BMI",
    "SleepHours",
    "PhysicalHealthDays",
    "MentalHealthDays",
    "PhysicalActivities",
    "HadDiabetes_binary",
    "HadHeartAttack",
    "HadStroke",
    "HadAngina"
]

df_sel = df[cols].dropna()

corr = df_sel.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)
plt.title("Heatmap de correlaciones entre variables de salud")
plt.show()
```

## NOTAS
Aqu podemos ver las variables que van ms relacionadas y por ende las que ms sentido tiene comparar.  
Ignoramos los "1.00" porque son la misma variable.

Que podemos observar:

- Los problemas cardiovasculares en general van bastate relacionados  
  con los ataques al corazn (en especial la angina de pecho).

- La diabetes por su lado tambin se relaciona con los ataques al corazn.

- Un IMC elevado (sobrepeso u obesidad) se ve ms relacionado con mayores problemas de  
  salud fsica y sobretodo con la presencia de diabetes.

- A su vez, los das de mala salud fsica y mental se ven bastante relacionados.

#### Conclusin
este grfico nos permite relacionar las variables ms importantes, observando las  
que se ven ms relacionadas claramente entre s.

```python
patron1 = (
    df.groupby(["GeneralHealth", "SleepHours"])["HadHeartAttack"]
    .mean()
    .reset_index()
)

patron1.head()

plt.figure(figsize=(12, 6))

sns.barplot(
    data=patron1,
    x="SleepHours",
    y="HadHeartAttack",
    hue="GeneralHealth"
)

plt.title("Proporcin de infartos segn salud general y horas de sueo")
plt.ylabel("Proporcin de personas con infarto")
plt.xlabel("Horas de sueo")
plt.legend(title="Salud general")
plt.show()
```

## NOTAS
En este grfico observamos las siguientes conclusiones:

- Existe una relacin clara entre la salud general y la proporcin de infartos:  
  a peor salud general (niveles ms bajos en la escala), mayor es la proporcin  
  de personas que han sufrido un infarto, independientemente de las horas de sueo.

- Las horas de sueo extremas se asocian con mayor riesgo: dormir muy  
  pocas horas (4 o menos) o demasiadas (9 o ms) tiende a mostrar proporciones  
  ms altas de infarto frente a dormir un nmero intermedio de horas.

- Dormir entre 6 y 8 horas parece estar asociado con menores  
  proporciones de infarto, especialmente en personas con mejor salud general.

- El sueo no acta de forma aislada: incluso con horas de sueo adecuadas,  
  una mala salud general mantiene un riesgo relativamente alto.

#### Conclusin
En resumen, el grfico refuerza que una buena salud general y un patrn de sueo  
moderado se asocian con menor proporcin de infartos, mientras que la mala salud  
y los extremos de sueo aumentan el riesgo.


```python
patron2 = (
    df.groupby(["SmokerStatus", "ECigaretteUsage"])["HadHeartAttack"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=patron2,
    x="SmokerStatus",
    y="HadHeartAttack",
    hue="ECigaretteUsage"
)

plt.title("Infarto segn tabaquismo y uso de cigarrillo electrnico")
plt.ylabel("Proporcin de infartos")
plt.xlabel("Estado de fumador")
plt.show()
```

## Notas
En este caso, el grfico nos presenta las siguientes conclusiones:

- Fumar se asocia con una mayor proporcin de infartos: a medida que el  
  estado de fumador aumenta (de no fumador a fumador ms intenso), la  
  proporcin de infartos crece de manera clara.

- El uso de cigarrillo electrnico no elimina el riesgo: incluso entre  
  quienes usan cigarrillo electrnico, la proporcin de infartos sigue  
  siendo relevante, especialmente en fumadores.

- La combinacin de tabaquismo y uso de e-cigarrillo muestra  
  los valores ms altos: los fumadores que adems usan cigarrillo  
  electrnico presentan las mayores proporciones de infarto.

- No fumadores y no usuarios de e-cigarrillo presentan el menor riesgo, lo  
  que refuerza el efecto negativo del consumo de nicotina sobre la salud cardiovascular.

#### Conclusin
En conjunto, el grfico sugiere que tanto el tabaquismo tradicional como el uso de cigarrillos  
electrnicos estn asociados con un mayor riesgo de infarto, y que combinarlos puede agravar an ms dicho riesgo.


```python
# HIPTESIS 2

patron3 = (
    df.groupby(["Sex", "AgeCategory"])["HadHeartAttack"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=patron3,
    x="AgeCategory",
    y="HadHeartAttack",
    hue="Sex"
)

plt.title("Problemas cardiacos en funcin del sexo y la edad")
plt.ylabel("Proporcin de infartos")
plt.xlabel("Edades")
plt.xticks(rotation=15)
plt.show()
```

## Notas
De forma breve, este grfico permite extraer estas conclusiones:

- La proporcin de infartos aumenta claramente con la edad en ambos sexos,  
  siendo muy baja en edades jvenes y creciendo de forma progresiva a partir de los 4550 aos.

- Los hombres presentan una mayor proporcin de infartos que las mujeres en  
  todos los grupos de edad, y la diferencia se hace ms marcada a partir de la mediana edad.

- En edades avanzadas (70 aos o ms) el riesgo aumenta de manera notable en  
  ambos sexos, aunque sigue siendo ms alto en hombres.

- Las mujeres muestran un incremento ms tardo, lo que sugiere  
  un posible efecto protector en edades ms jvenes.

#### Conclusin
Dicho esto, el grfico indica que la edad es un factor de riesgo clave para los infartos,  
y que el sexo masculino presenta un riesgo sistemticamente mayor, especialmente en edades avanzadas.
